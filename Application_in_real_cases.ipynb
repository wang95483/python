### make changes in data frame

* .apply() <br>
`axis = 0` – apply function to each column (variables) <br>
`axis = 1` – apply function to each row (observations) <br>
axis = 0 is the default parameter

````python
df_subset.apply(np.max, axis=0)
````

````python
df[['country','year', 'rgdpo','pop']].apply(np.max, axis=0)
````

.apply() can be some built-in functions like the max function, a lambda function, or a user-defined function

````python
def square(x):
    return x * x

df['rgdpo'].apply(square)
````

````python
#add a new column (variable) in the data frame
df['rgdposquare']=df['rgdpo'].apply(lambda x: x * x)
df[1300:1320]
````

* df.where() <br>
“keep” the rows we have selected and replace the rest rows with any other values

````python
df.where(df['pop'] >= 2, False)
````

* .loc[] <br>
specify the column that we want to modify and assign values

````python
df.loc[df['rgdpo'] == max(df['rgdpo'])]
````

````python
df.loc[df['year'] == max(df['year'])] = np.NaN
df
````

* .apply() <br>
modify rows/columns as a whole

````python
def update_row(row):
    # modify pop
    row['pop'] = row['pop']*1000000
    return row
````

````python
df.apply(update_row, axis=1)
````

* .map() method <br>
modify all individual entries in the dataframe altogether

````python
df.map(lambda x : round(x,2) if type(x)!=str else x) #changethe nunber of decimal
````

`np.isnan()` is used in NumPy to check whether elements in an array or dataset are NaN (Not a Number) values.

````python
#replacing missing values
def replace_nan(x):
    if type(x)!=str:
        return 0 if np.isnan(x) else x
    else:
        return x
````

````python
df.map(replace_nan)
````

### Data visualization with dataframes

`df.country.isin()`: filter a DataFrame by checking whether the values in a column exist in a given list. <br>
```python
#Syntax
df['column_name'].isin([list_of_values])

````python
#select the range we want
df1 = df[(df.country.isin(['Argentina', 'India', 'South Africa', 'United States']))
& (df['year'] == 2010) ]
#show it
df1
````

```python
#set the index
df1.set_index('column_name')

````python
#set the index to the subset of the data frame
df1 = df1.set_index('country')
df1
````

````python
#give the columns slightly better names
df1.columns = 'year','population', 'total GDP'
df1
````

````python
#change units
df1['population'] = df1['population'] *1e6 #here, population variable is in millions, let’s revert to single units
df1
````

````python
#add a column showing real GDP per capita (GDP and pop are in diﬀerent units)
df1['GDP percap'] = df1['total GDP']*1e6/df1['population']
df1
````

`govt_debt.index.droplevel(-1)` <br>
`govt_debt.index` $\rightarrow{}$ Accesses the index of the DataFrame. <br>
`.droplevel(-1)` $\rightarrow$ Drops the last level of a MultiIndex (since -1 refers to the last level).<br>

````python
ind = govt_debt.index.droplevel(-1)
govt_debt.index = ind
govt_debt
````
